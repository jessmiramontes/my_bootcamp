```python
#si yo tengo una muestra puedo estimar la media usando bootstraping
#La técnica de bootsraping es simular que la muestra tiene suficiente información de la distribución subyacente 
#y se aplica haciendo un resampling, muestreo con remplazo, para aproximar los parámetros de la población

#Los intervalos de confianza son Improbalbe 10% es z=1.65, Improbable 5% z=1.96, Improbable 1% z= 2.58 
```

```python
#intervalo de confianza 2010
mean_2010 = accidents2010.mean()
std2010 = accidents2010.std()
n2010 = len(accidents2010)
#para el 95% de datos
ci = [mean_2010 - std2010 *1.96 / np.sqrt(n2010), mean_2010 + std2010 *1.96 / np.sqrt(n2010)]
ci
```

```python
#hay una fábrica que produce 150 unidades por hora de un producto a, con una desviación estándar de 8 unidades
#dicen que hay una máquina que hace lo mismo que hace 160 unidades por hora con una desviación estándar de 8 unidades
#¿compro la nueva máquina o no?
#total de muestras es igual a 35
#nivel alfa de significancia 0.01, imaginando que la máquina es muy cara

z_score = (160 - 150) / (8/np.sqrt(35))
z_score
```

```python
p_value = 2 * (1-stats.norm.cdf(abs(z_score)))
p_value
```

```python
#un automóvil a 30 millas por hora debe recorrer máximo 65 pies al frenar 
#un fabricante de automóviles ha diseñado un nuevo sistema de frenos que recorre una distacia de 63.5 pies con una
#desviación estándar de 4 pies habiéndolo probado en 64 autos
#alfa 0.01 por un tema de seguridad

z_score = (63.5 - 65) / (4/np.sqrt(6))
p_value = (1-stats.norm.cdf(abs(z_score)))
print(z_score)
print(p_value)
```
