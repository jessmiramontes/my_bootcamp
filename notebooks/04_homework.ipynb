### *Ejercicio 1*
Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
Divide el número del paso 2 entre el número de iteraciones(1000).

```python
#probabilidad de que la suma de los dos dados me de 2
#generar contenido aleatorio
import random
#simulación lanzamiento de moneda
table = 0
for i in range(100000):
    result = random.randint(0,1)
    if result==0:
        table = table + 1
print(table/100000)
```

```python
#simular lanzamiento de dado y la probabilidad de que sea par
table = 0
iterations = 10000
for i in range(iterations):
    result = random.randint(1,6)
    if result in [1,2]:
        table = table + 1
print(table/iterations)
```

```python
#probabilidad de sacar un joker o una queen
table = 0
iterations = 10000
for i in range(iterations):
    result = random.randint(1,13)
    if result in [11,12]:
        table = table + 1
print(table/iterations)
```

```python
#calcular la probabilidad de sacar 2 caras en un lazamiento de monedas
coin_1= ["H", "T"]
coin_2= ["H", "T"]
table = 0
iterations = 100000
for i in range(iterations):
    result_1 = random.choice(coin_1)
    result_2 = random.choice(coin_2)
    if result_1 in ["H"] and result_2 in ["H"]:      
        table = table + 1
print(table/iterations)
```

```python
#Ejercicio 1
dice_1= [1,2,3,4,5,6]
dice_2= [1,2,3,4,5,6]
table = 0
iterations = 1000
for i in range(iterations):
    result_1 = random.choice(dice_1)
    result_2 = random.choice(dice_2)
    result = result_1 + result_2
    if result > 7 or result % 2==0:      
        table = table + 1
print(table/iterations)
```
